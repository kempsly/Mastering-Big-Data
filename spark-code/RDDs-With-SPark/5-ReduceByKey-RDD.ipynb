```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("distinct")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
rdd = sc.textFile('sample.txt')
rdd.collect()
```

```python
rdd2 = rdd.flatMap(lambda x: x.split(' '))
```

```python
rdd3 = rdd2.map(lambda x: (x,1))
rdd3.collect()
```

```python
rdd3.reduceByKey(lambda x,y: x+y).collect()
```

# Worcount Program

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("WordcountExample")
sc = SparkContext.getOrCreate(conf=conf)
```

```python

rdd = sc.textFile('playground.txt')
```

```python


rdd2 = rdd.flatMap(lambda x: x.split(' '))
rdd2.collect()
```

```python

rdd3 = rdd2.filter(lambda x: len(x) != 0)
```

```python
rdd3.collect()
```

```python
rdd4 = rdd3.map(lambda x: (x,1))

rdd4.collect()
```

```python
rdd4.reduceByKey(lambda x,y : x+y).collect()
```

```python



rdd.flatMap(lambda x : x.split(' ')).map(lambda x: (x,1)).reduceByKey(lambda x,y : x+y).collect()
```
