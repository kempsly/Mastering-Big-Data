```python
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("Grouping")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
rdd = sc.textFile('sample_columned.txt')
rdd.collect()
```

```python
rdd2 = rdd.flatMap(lambda x: x.split(" "))
rdd3 = rdd2.distinct()
rdd3.collect()
```

```python
rdd_new = sc.textFile('playground.txt')
rdd_new.collect()
```

```python
rdd_new_filtered = rdd_new.filter(lambda x: x !="")
rdd_new_filtered.collect()
```

```python
rdd_new_filtered.map(lambda x: (x, len(x.split(' ')))).collect()
```

```python
rdd_new_filtered.flatMap(lambda x: (x, len(x.split(' ')))).collect()
```

```python
rdd_new2 = rdd_new.flatMap(lambda x: x.split(' '))
rdd_new2.collect()
```

```python
rdd_new2.map(lambda x: (x, len(x))).collect()
```

```python
rdd_new2.map(lambda x: (x, 1)).collect()
```

```python
rdd_new3 = rdd_new2.map(lambda x: (x, len(x)))
rdd_new3.collect()
```
