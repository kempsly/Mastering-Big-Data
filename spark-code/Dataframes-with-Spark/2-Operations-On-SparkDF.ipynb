## Manipulate columns from Spark DF using WithColumns function

The `withColumn` function in PySpark is used to either create a new column or update an existing column in a DataFrame. It allows you to apply transformations or computations on the column data.

---

### **Syntax**
```python
DataFrame.withColumn(colName, col)
```

- **`colName`**: The name of the column to be added or updated.
- **`col`**: The new column expression, which can be created using PySpark functions.

---

### **Common Use Cases**
1. **Add a New Column**:
   Add a column based on computations or transformations.
2. **Update an Existing Column**:
   Replace the existing column with new values.
3. **Apply Transformations**:
   Use functions like `lit()`, `when()`, or custom expressions.

---




````python
df.printSchema()
````

````python
#Type casting : changing the type of roll from integer to string
df3 = df.withColumn("roll", col("roll").cast("String"))
````

````python
df3.printSchema()
````

````python
df3.show()
````

````python
#Adding 10 to each value in the marks column
df3 = df3.withColumn('marks', col('marks') + 10)
df3.show()
````

````python
#Create a new column in the dataframe called aggregated marks
df3 = df3.withColumn('aggregatted marks', col('marks') - 10)
df3.show()
````

````python
from pyspark.sql.functions import col, lit
df3 = df3.withColumn("name", lit("USA"))
df3.show()
````

````python
df3 = df3.withColumn("marks", col("marks") - 10).withColumn("updated marks", col("marks") + 70).withColumn("Country", lit("USA"))
df3.show()
````

## Spark User Defined Functions

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf
from pyspark.sql.types import IntegerType
spark = SparkSession.builder.appName("SparkUDF").getOrCreate()
````

````python
df = spark.read.options(header='True', inferSchema='True').csv('data_office.csv')
df.show()
````

````python
df.show()
````

````python

def get_total_salary(salary):
  return salary + 100
````

````python
totalSalaryUDF = udf(lambda x: get_total_salary(x), IntegerType())
````

````python
totalSalaryUDF
````

````python
df.withColumn("total_salary", totalSalaryUDF(df.salary)).show()
````

````python
from pyspark.sql.types import DoubleType
````

````python
def get_incr(state, salary, bonus):
  sum = 0
  if state == "NY":
    sum = salary * 0.10
    sum += bonus * 0.05
  elif state == "CA":
    sum = salary * 0.12
    sum += bonus * 0.03
  return sum
````

````python
incrUDF = udf(lambda x,y,z: get_incr(x,y,z), DoubleType())

df.withColumn("increment", incrUDF(df.state, df.salary, df.bonus)).show()
````
